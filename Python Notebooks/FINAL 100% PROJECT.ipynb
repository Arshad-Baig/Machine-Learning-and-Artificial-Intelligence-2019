```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.preprocessing import scale
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('DataFiles/CarsPrices.csv')
```

```python
data.tail()
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dataFrame=data
dataFrame.make=le.fit_transform(data.make)
dataFrame.fuel_type=le.fit_transform(data.fuel_type)
dataFrame.aspiration=le.fit_transform(data.aspiration)
dataFrame.num_doors=data.num_doors.astype(str)
dataFrame.num_doors=le.fit_transform(data.num_doors)
dataFrame.body_style=le.fit_transform(data.body_style)
dataFrame.drive_wheels=le.fit_transform(data.drive_wheels)
dataFrame.engine_location=le.fit_transform(data.engine_location)
dataFrame.engine_type=le.fit_transform(data.engine_type)
dataFrame.num_cylinders=le.fit_transform(data.num_cylinders)
dataFrame.fuel_system=le.fit_transform(data.fuel_system)
dataFrame.head()
```

```python
dataFrame.dtypes
```

```python
dataFrame=dataFrame.replace('?',np.nan)
dataFrame
```

```python
dataFrame.mean()
```

```python
from sklearn.preprocessing import Imputer
```

```python
imp=Imputer(missing_values="NaN",strategy="mean",axis=0)
imp=imp.fit(dataFrame.iloc[:,:])
dataFrame.iloc[:,:]=imp.transform(dataFrame.iloc[:,:])
dataFrame.head()
```

```python
dataFrame=dataFrame.astype(int)
```

```python
dataFrame.dtypes
```

```python
x=dataFrame.iloc[:,0:26]
y=dataFrame.price
```

```python
x.head()
```

```python
y.head()
```

```python
x=scale(x)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x=pd.DataFrame(x)
x.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn=KNeighborsClassifier(n_neighbors=7,metric='euclidean')
model_knn.fit(x_train,y_train)
```

```python
model_knn
```

```python
y_predict=model_knn.predict(x_test)
print(len(y_predict))
y_predict
```

```python
np.mean(y_predict)
```

```python
np.mean(y_test)
```

```python
(y_predict==y_test)
```

```python
accuracy_score(y_predict,y_test)
```
