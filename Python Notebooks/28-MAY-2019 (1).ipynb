### Problem 1 :

#### Problem Statement :
Print no. of matches won by each team based on year wise (each year as one column)

```python
import pandas as pd
df=pd.read_csv('DataFiles/Match.csv')

unique_teams=[ ]
for i in range(len(df.values)):
    if df.values[i][2] not in unique_teams:
        unique_teams.append(df.values[i][2])

unique_years=[ ]
for i in range(len(df.values)):
    if df.values[i][5] not in unique_years:
        unique_years.append(df.values[i][5])
        
count=0
df_result=pd.DataFrame()
df_result['Teams']=unique_teams
for i in unique_years:
    array=[ ]
    for j in unique_teams:
        count=0
        for k in range(len(df.values)):
            if(i==df.values[k][5] and j==df.values[k][10]):
                count=count+1
        array.append(count)
    df_result[i]=array
df_result
```

### Problem 2 :

#### Problem Statement :
Print  ManOfMach count of each player in  Hyderabad stadium

```python
import pandas as pd
import numpy as np
df=pd.read_csv('DataFiles/Match.csv')

mom=[ ]
for i in range(len(df.values)):
    mom.append(df.values[i][14])

hyd_mom_players=[ ]
for i in range(len(df.values)):
    if(df.values[i][7]=='Hyderabad' and df.values[i][14]==mom[i]):
        hyd_mom_players.append(df.values[i][14])
        
unique_hyd_mom_players=[ ]
for i in range(len(hyd_mom_players)):
    if hyd_mom_players[i] not in unique_hyd_mom_players:
        unique_hyd_mom_players.append(hyd_mom_players[i])
unique_hyd_mom_players.sort()

no_of_matches=[ ]
for i in range(len(unique_hyd_mom_players)):
    no_of_matches.append(hyd_mom_players.count(unique_hyd_mom_players[i]))

df1=pd.DataFrame()
df1['Players']=unique_hyd_mom_players
df1['No.of MoMs in Hyderabad']=no_of_matches
df1
```

### Problem 3 :

#### Problem Statement :
Print number of matches won and loss(as two columns). Consider as win when a team wins Toss  and Match . print for each and every team.

```python
df=pd.read_csv('DataFiles/Match.csv')
teams=[ ]
for i in range(len(df)):
    if(df.values[i][2] not in teams):
        teams.append(df.values[i][2])
#teams
w_count=l_count=0
won_count=[]
loss_count=[ ]

for i in teams:
    w_count=l_count=0
    for j in range(len(df)):
        if(i==df.values[j][9] and i==df.values[j][10]):
            w_count=w_count+1
        
        elif(i==df.values[j][2] or i==df.values[j][3]):    #at first i put df.values[j][9] and df.values[j][10]
            l_count=l_count+1
            
    won_count.append(w_count)    
    loss_count.append(l_count)
df2=pd.DataFrame()
df2['Teams']=teams
df2['No.of Times Won']=won_count
df2['No.of Times Lost']=loss_count
df2
```
