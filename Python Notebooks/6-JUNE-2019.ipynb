### Using fillna

```python
data.fillna(0)              # fillna fills the Nan values with '0' provided inside the bracket
```

```python
data.fillna(method="ffill")
# this fills the 'NaN' values with the above value of that entry
```

```python
data.fillna(method="bfill")
# this fills the 'NaN' values with the below values of that entry
```

```python
data
```

```python
data.fillna(method="ffill",axis='columns')  #   axis is used to denote the  ffill value to be taken from 'rows' or 'columns'
```

```python
data.fillna(method="bfill",axis='columns')
```

### Using interpolate

```python
data.interpolate()  # interpolate fills the 'NaN' by guessing it based on upper and lower values
```

### Using dropna

```python
data
```

```python
data.dropna()  # drops all the rows where there is a 'NaN'
```

```python
data.dropna(how="all")  # drops the rows only if all the values in that row are 'NaN'..else it does not drop any row
```

```python
data.dropna(how="any")  # drops the rows even if any one value in that row is 'NaN'..else it does not drop any row
```
